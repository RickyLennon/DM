```python
filas, columnas = df_ordenado.shape

Pertenencia = np.zeros((filas))
Neuronas = np.zeros((3, columnas))

Neuronas[0] = valores_mÃ­nimos
Neuronas[1] = valores_mÃ¡ximos
Neuronas[2] = valores_promedio

Distancia = np.zeros((1, 3))
Base = np.array(df_ordenado)
```

```python
for k in range(3):
    
    Cluster = np.zeros((1, 3))  

    for i in range(filas):
        
        for j in range(3):
            
            Distancia[0, j] = np.linalg.norm(Base[i, :] - Neuronas[j, :])

        un = 0.5 * (1 - i / filas)

        min_dist = Distancia.min()
        
        if (Distancia[0, 0] == min_dist):
            Neuronas[0, :] = Neuronas[0, :] + (un * (Base[i, :] - Neuronas[0, :]))
            Cluster[0, 0] += 1
            Pertenencia[i] = 1
        else:
            if (Distancia[0, 1] == min_dist):
                Neuronas[1, :] = Neuronas[1, :] + (un * (Base[i, :] - Neuronas[1, :]))
                Cluster[0, 1] += 1
                Pertenencia[i] = 2
            else:
                Neuronas[2, :] = Neuronas[2, :] + (un * (Base[i, :] - Neuronas[2, :]))
                Cluster[0, 2] += 1
                Pertenencia[i] = 3
```

```python
Info3 = {
    "Neurona 1": Cluster[0,0],
    "Neurona 2": Cluster[0,1],
    "Neurona 3": Cluster[0,2]
}
Clusters = pd.DataFrame(Info3, index=[0]) 
Clusters.index = Clusters.index + 1
Clusters
```
