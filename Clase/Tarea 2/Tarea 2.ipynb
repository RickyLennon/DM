```python
import os
import sys
import pyspark
import random
import time
import multiprocessing

import findspark
findspark.init()
findspark.find()

from pyspark.sql import SparkSession
from pyspark.sql import DataFrame, SparkSession, Window
from pyspark.sql.functions import row_number, max, min, col
import pyspark.sql.functions as F
```

```python
Spark = SparkSession.builder.appName("Tarea 2").getOrCreate()
nombre_app = Spark.sparkContext.appName
print(f"El nombre de la aplicación es: {nombre_app}")
```

```python
df = Spark.read.csv("ENOE.csv", header=True, inferSchema=True)
```

```python
window_spec = Window.orderBy("ent")
df_with_index = df.select("ent", "sex", "eda", "clase1", "e_con", "n_hij",  "ingocup").withColumn("index", row_number().over(window_spec))
columns = ["index", "ent", "sex", "eda", "clase1", "e_con", "n_hij",  "ingocup"]
df_with_index = df_with_index.select(columns)
df_with_index.show()
total_individuos = df_with_index.select("index").count()
print(f"Total individuos: {total_individuos}")
```

```python
rdd_from_df = df_with_index.rdd

filtered_rdd = rdd_from_df.filter(lambda x: x["clase1"] == 1 and x["n_hij"] != None and x["ingocup"] != None)
filtered_rdd2 = filtered_rdd.filter(lambda x: x["ent"] == 9 or x["ent"] == 19) 
filtered_rows = filtered_rdd2.collect()
filtered_rows
```

```python
filtered_rdd2.count()
```

```python
columns = df.select("ent", "sex", "eda", "clase1", "e_con", "n_hij", "ingocup")
df_filtered = columns.filter((F.col("clase1") == 1) & (F.col("n_hij").isNotNull()) & (F.col("ingocup").isNotNull()))
df_filtered2 = df_filtered.filter((F.col("ent") == 9))
df_filtered2.show()
```

```python
#Número de hijos promedio que tiene la PEA en Nuevo León 
Nuevo = df_filtered2.select("n_hij").rdd #Convierte la filtración anterior a RDD
resultado = Nuevo.flatMap(lambda x: x) #Convierte elementos columna en elementos fila
resultado2 = resultado.reduce(lambda x, y: x + y)
Promedio = resultado2/resultado.count()
Promedio
```

```python
Filtro = df_with_index.filter((F.col("clase1") == 1) & (F.col("n_hij").isNotNull()) & (F.col("ingocup").isNotNull()))
Filtro2 = Filtro.filter((F.col("n_hij") == 2) & (F.col("ent") == 9))
Filtro2.show()
```

```python
df_filtered3 = df_filtered.filter((F.col("ent") == 19))
df_filtered3.show()
```

```python
#Número de hijos promedio que tiene la PEA en Nuevo León 
Nuevo2 = df_filtered3.select("n_hij").rdd #Convierte la filtración anterior a RDD
resultado3 = Nuevo2.flatMap(lambda x: x) #Convierte elementos columna en elementos fila
resultado4 = resultado3.reduce(lambda x, y: x + y)
Promedio2 = resultado4/resultado3.count()
Promedio2
```
